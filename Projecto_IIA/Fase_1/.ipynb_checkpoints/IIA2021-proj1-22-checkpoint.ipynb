```python
from searchPlus import* 
from copy import copy,deepcopy
class PuzzleColares(Problem):
    pulseiras={ "1":["g","g","g","g","g","b","g","g","p","p","r","r","r","r","r","r","r","r","r","g"],
           "2":["b","b","b","b","b","b","b","p","p","p","p","p","p","g","g","p","p","g","g","b"]}
    obj={ "1":["g","g","g","g","g","g","g","p","r","r","r","r","r","r","r","r","r","g","g","g"],
           "2":["b","b","b","b","b","b","b","p","p","p","p","p","p","p","p","p","p","g","b","b"]}


    def __init__(self,initial=pulseiras,goal=obj):
        self.initial,self.goal=initial,goal #inicializacao atributos
        self.verificaInicial() #verifica se o colar esta na condicao inicial que deve
    
    #metodo que verifica se o colar inserido eh valido
    #este metodo verifica se as cores das posicoes de interseção de um anel sao iguais as do anel seguinte,
    # se o anel tem o tamanho correto e se a quantidade de missangas por cor eh valida
    def verificaInicial(self): 
        var=[]
        y=0
        differentColors=[] #variavel que vai receber as diferentes cores
        contador=[]
        conta9s=0 #conta quantos grupos de 9 elementos sao formados
        conta10s=0 #conta quantos grupos de 10 elementos sao formados
        for x in self.initial:
            var.append(self.initial[str(y+1)])                
            y+=1
        for x in range(len(var)-1) :
            if var[x][3]!=var[x+1][17] or var[x][7]!=var[x+1][13] or len(var[x])!=20: #verifica se os elementos do meio são iguais ou se o tamanho do anel e diferente de 20
                return False
        for x in var :
            for y in range(len(x)):
                if not differentColors.__contains__(x[y]): #verificar se uma dada cor esta na lista de cores
                    if (y==7 or y==3) and var.index(x)!=(len(var)-1): #nao pode contar a mesma cor duas vezes
                        pass
                    else:
                        differentColors.append(x[y])  #insere cor na lista differentColors
                        contador.insert(differentColors.index(x[y]),1)
                elif (y==7 or y==3) and var.index(x)!=len(var)-1:
                    pass
                else:
                    contador[differentColors.index(x[y])]+=1
                    
        for x in contador: #vai verificar quantos conjuntos de 9 e 10 elementos existem
            if int(x)==10:
                conta10s+=1
            if int(x)==9:
                conta9s+=1
        return conta9s==len(contador)-2 and conta10s==2
    
    #define resultado de uma acao num dado estado
    #face a acao inserida, o metodo movimenta o anel em questao no sentido selecionado
    def result(self, estado,accao):
        resultado=deepcopy(estado) #inicializacao do dicionario que vai receber o caminho-resultado
            
        accoes=self.actions(estado) #chamada do metodo actions para listar as varias acoes a partir do estado
        if accao in accoes:
            sentido=accao.replace(accao[0], "") #guarda o sentido H ou aH
            anel=accao[0] #guarda o num anel que vai rodar
            anelAnt=str(int(anel)-1) #anel anterior ao selecionado
            anelSeg=str(int(anel)+1) #anel seguinte ao selecionado
            if sentido=="H": #se o sentido for horario, as missangas que estavam num sitio avancam uma casa para a direita
                for x in range(len(resultado[anel])): #percorre os varios elementos do anel
                    if x==0:
                        atual=resultado[anel][x]
                    if x< len(resultado[anel])-1:
                        proximo=resultado[anel][x+1] #avancar de elementos uma vez para a direita
                        resultado[anel][x+1]=atual
                        atual=proximo
                    else:
                        resultado[anel][0]=atual        
            
            else: #se o sentido for antihorario, a missanga num dado sitio avanca uma casa para a esquerda
                for x in range(len(resultado[anel])):
        
                    if x!= len(resultado[anel])-1:  
                        resultado[anel][x]=resultado[anel][x+1] #avanco para a esquerda
                    else:
                        resultado[anel][x]=estado[anel][0]   
           
                if list(resultado).index(anel)==0: #alteracao das missangas do meio
                    resultado[anelSeg][13]=resultado[anel][7] 
                    resultado[anelSeg][17]=resultado[anel][3]
                
                elif list(resultado).index(anel)==len(resultado)-1:
                    resultado[anelAnt][3]=resultado[anel][17] 
                    resultado[anelAnt][7]=resultado[anel][13]
                
                else:
                    resultado[anelAnt][3]=resultado[anel][17] 
                    resultado[anelAnt][7]=resultado[anel][13]
                    resultado[anelSeg][13]=resultado[anel][7] 
                    resultado[anelSeg][17]=resultado[anel][3]
                
            
        return resultado
    #define acoes possiveis para um dado estado 
    #devolve as acoes passiveis de ser realizadas, face ao numero de aneis 
    def actions(self,estado): 
        teste=estado.keys()
        accoes=[] #inicializacao da lista que vai receber as acoes 
        for x in teste:
            accoes.append(x+"H") #girar no sentido dos ponteiros do relogio
            accoes.append(x+"aH") #girar no sentido contrario aos ponteiros do relogio
        return accoes
    
    #verifica se um dado estado e o desejado(goal), isto e, se todas as posicoes de cada anel apresentam uma cor igual ah cor
    #da mesma posicao no respetivo objetivo
    def goal_test(self,estado): 
        
        for anel in range(len(estado)):
            for element in range(len(estado[str(anel+1)])):
                if estado[str(anel+1)][element]!=self.goal[str(anel+1)][element]:  
                    return False
        return True
    
    #determina custo que tem um estado ao realizar uma acao para prosseguir para um dado estado
    def path_cost(self, cost,state,action,new_state):
        
        return cost+1
    
    #visualizacao grafica dos aneis; permite a visualizacao do estado dos aneis no terminal
    def display(self,estado): 
        row=9
        matriz = [["" for i in range(9+6*(len(self.initial)-1))] for j in range(9)] #construcao da matriz de aneis que permita a visualizacao dos aneis

        #como os aneis estao lado a lado, o num de linhas nao muda, mas sim o de colunas logo tera smp 9 linhas e o valor acima 
        #colocado para as colunas
        
        col=9+6*(len(self.initial)-1)
        
        
        for i in range(0,row):
            for j in range(0,col):
                if i==0 or i==8:
                    if j%3==0 and j%6!=0  and j>0 and j<len(matriz[0]):
                        anel=int(j/6)+1
                        if i==0:
                            matriz[i][j]=estado[str(anel)][19]
                            matriz[i][j+1]=estado[str(anel)][0]
                            matriz[i][j+2]=estado[str(anel)][1]
                        else:
                            matriz[i][j]=estado[str(anel)][11]
                            matriz[i][j+1]=estado[str(anel)][10]
                            matriz[i][j+2]=estado[str(anel)][9]
                        
                elif i== 1 or i==7:
                    if (j==2 or j<len(matriz[0])-1 ) and ((j!=0 and (j-2)%6==0)) :
                        anel=int(j/6)+1
                        if i==1:
                            matriz[i][j]=estado[str(anel)][18]
                        else:
                            matriz[i][j]=estado[str(anel)][12]
          
                    elif (j!=0 and j%6==0):
                        anel=int(j/8)+1
                        if i==1:
                            matriz[i][j]=estado[str(anel)][2]
                        else:
                            matriz[i][j]=estado[str(anel)][8]

                        
                elif i==3:
                    if (((j==0 or j%6==0) and j<len(matriz[0])-3)):
                        anel=int(j/6)+1
                        matriz[i][j]=estado[str(anel)][16]
                        matriz[i+1][j]=estado[str(anel)][15]
                        matriz[i+2][j]=estado[str(anel)][14]
                        
                    elif (j==8 or  (j>8 and (j-2)%6==0)):
                        if j==8:
                            anel="1"
                        else:
                            anel=int(j/6)
                        matriz[i][j]=estado[str(anel)][4]
                        matriz[i+1][j]=estado[str(anel)][5]
                        matriz[i+2][j]=estado[str(anel)][6]
                        
                        
                        
                        
                elif i==2 :
                    if (j-1)%6 ==0:
                        if j<len(matriz[0])-2:
                            anel=int((j-1)/6)+1
                            matriz[i][j]=estado[str(anel)][17]

                        else:
                            anel=int((j-1)/6)
                            matriz[i][j]=estado[str(anel)][3]
                elif i==6:
                    if (j-1)%6 ==0:
                        if j<len(matriz[0])-2:
                            anel=int((j-1)/6)+1
                            matriz[i][j]=estado[str(anel)][13]
                        else:
                            anel=int((j-1)/6)
                            matriz[i][j]=estado[str(anel)][7]
                   
 
        
                
        
        print('\n'.join(["".join(['{:2}'.format(item) for item in row]) #impressao propriamente dita
        for row in matriz]))        

    
        
    def exec(p,estado,accoes):
        """ Executa uma sequência de acções a partir do estado
        devolve um par (estado, custo) depois de imprimir
        """
        custo = 0
        for a in accoes:
            seg = p.result(estado,a)
            custo = p.path_cost(custo,estado,a,seg)
            estado = seg
        return (estado,custo)
    
#criacao de uma instancia da classe para possibilitar a execucao do programa
p=PuzzleColares() #NOTA:apesar de ser utilizado o construtor base, o codigo permite a utilizacao de um numero de aneis superior a 2!
#determinacao do resultado da acao de chamar uma vez o antihorario no estado inicial
p.result(p.initial, "1aH")
#realiza o conjunto de acoes possiveis, face aos aneis disponiveis;se estiver bem feito o resultado sera o estado inicial
#uma vez que, se cada anel se pode movimentar de forma H e aH,a rotacao de cada um sera sempre anulada
a=p.exec(p.initial, p.actions(p.initial))
#mostrar a visualizacao do resultado dos colares apos o metodo exec
p.display(list(a)[0])
```
